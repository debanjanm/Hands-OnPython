Problem (brief): Given an array nums, move all 0's to the end while maintaining the relative order of the non-zero elements. Do this in-place with minimal operations.

```python
## Solution 1:

def moveZeroes(nums: list[int]) -> None:
    """
    Modifies nums in-place to move all zeros to the end while preserving order.
    """
    j = 0  # index to place next non-zero
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[j] = nums[i]
            j += 1
    # fill remaining with zeros
    for k in range(j, len(nums)):
        nums[k] = 0
```

```python
# Example
arr = [0,1,0,3,12]
moveZeroes(arr)
print(arr)  # [1, 3, 12, 0, 0]
```
